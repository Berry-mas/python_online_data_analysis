```python
import time

from selenium import webdriver # selenium의 webdriver 모듈을 임포트 합니다.
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

# chromedriver.exe 파일의 경로를 입력합니다.
s = Service(r'C:\python_onlinedata\chromedriver-win64\chromedriver.exe')
driver = webdriver.Chrome(service=s)  #크롬 브라우저 실행

url = 'https://www.yes24.com/Product/Goods/104663596'   #웹페이지로 이동 (접속)
time.sleep(1)
driver.get(url) 

#베스트 버튼 찾아서 클릭
element = driver.find_element(by=By.XPATH, value='//*[@id="yesFixCorner"]/dl/dd/ul[1]/li[1]/a') # xpath link of the elementt
element.click()

#1위 책 클릭 (에일리언 스테이지 아트북 특별판)
element = driver.find_element(by=By.XPATH, value='//*[@id="yesBestList"]/li[1]/div/div[2]/div[2]/a[1]') # xpath link of the elementt
element.click()

html = driver.page_source
time.sleep(3)
driver.close()
```
