# 1. 기사

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows 사용자
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS 사용자
# plt.rcParams['font.family'] = 'NanumGothic'  # Linux 사용자

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
import os
import numpy as np
from os import listdir
from os.path import isfile, join

# 여러 디렉토리 경로
mypaths = ['./윤석열 기사 200_연합뉴스/', './윤석열 기사 200_조선일보/', './윤석열 기사 200_한겨레/']

# 모든 파일을 저장할 리스트
total_docs = []

# 각 디렉토리를 순회
for mypath in mypaths:
    onlyfiles = [f for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]
    onlyfiles.sort()

    # 파일 내용 읽기
    for file in onlyfiles:
        file_path = os.path.join(mypath, file)
        with open(file_path, 'r', encoding='utf8') as f:
            content = f.read()
        total_docs.append(content)

# 결과 출력 (첫 번째 문서의 앞 100자 출력)
print(f"총 문서 수: {len(total_docs)}")
```

```python
from kiwipiepy import Kiwi
import re
kiwi = Kiwi()
from kiwipiepy.utils import Stopwords
stopwords_dict = Stopwords()

# 사용자 정의 불용어를 키위 형태소 분석기의 불용어 사전에 추가하기
customized_stopwords = [
    '취재', '뉴스', '연합', '조선', '한겨레', '사전', '조사', '서울', '사진', '대표', '중앙',
    '유세', '더불어민주당', '국민의힘', '더불어 민주당', '국민의 힘', '허경영', '문재인', '박근혜',
    '원희룡', '윤석렬', '단일', '철수', '토론', '대선', '정권', '나라', '지지', '교체', '생각', '인간',
    '정치', '기사제보', '연합뉴스', '재배포', '사진구매', '조선일보', '한겨레', '토트넘', '후보',
    '안철수', '윤석열', '이재명', '기자', '대통령', '광고', '이날', '오전', '오후', '관련', '이번',
    '시작', '당시', '가능', '이후', '이상', '공개', '자신', '당선', '시간', '시절', '이야기', '전날',
    '저녁', '지난해', '지금', '민주당', '국힘', '국민', '투표', '선거', '대한민국', '포함', '정도',
    '말씀', '해당', '일정', '일부', '하루', '발언', '지난달', '미래', '가운데', '경우', '부분',
    '정부', '국민의당', '의원', '포인트', '공동', '국민의 당', '경북', '제주', '호남', '영남', '대구', '대전',
    '경기', '한국', '부산', '충청', '광주', '전북', '전남', '충남', '성남', '만큼', '경남', '경기도', '심상정', '정의당'
]
for word in customized_stopwords:
    stopwords_dict.add((word, 'NNG'))
    stopwords_dict.add((word, 'NNP'))

def Kr_preprocessing(text):
    text = re.sub(r'[^\d\s\w]','',text)
    kiwi_tokens = kiwi.tokenize(text, stopwords=stopwords_dict)
    noun_words=[]
    for token in kiwi_tokens:
        if 'NN' in token.tag and len(token.form)>1:
            noun_words.append(token.form)
    # 길이가 1인 단어 제거 하기
    final_noun_words = []
    for word in noun_words:
        if len(word)>1:
            final_noun_words.append(word)
    return final_noun_words
```

```python
filtered_docs = [Kr_preprocessing(doc) for doc in total_docs]
```

```python
from itertools import chain
filtered_docs_one = list(chain.from_iterable(filtered_docs))
```

```python
len(filtered_docs_one)
```

```python
from collections import Counter
top = Counter(filtered_docs_one)
print(top.most_common(10))
```

```python
import matplotlib.pyplot as plt

# 상위 20개의 단어와 빈도만 사용
top_n = 20  # 표시할 단어 수
words, counts = zip(*top.most_common(top_n))  # 단어와 빈도를 분리

# 막대그래프 생성
plt.figure(figsize=(12, 6))
plt.barh(words, counts, color='#E61E2B', alpha=0.8)  # 수평 막대그래프
plt.xlabel('빈도수', fontsize=12)
plt.ylabel('단어', fontsize=12)
plt.title('상위 20 기사 빈도 분석 (검색어: 윤석열)', fontsize=16)
plt.gca().invert_yaxis()  # y축 뒤집기 (가장 빈도가 높은 단어가 위로 오도록)

# 그래프 출력 및 저장
plt.tight_layout()
output_file = 'ysy_top_20_words_analysis.png'  # 저장할 파일 이름
plt.savefig(output_file, dpi=300, bbox_inches='tight')  # DPI와 여백 설정
plt.show()
```

# 2. 댓글

```python
import os
import random

# 디렉토리 경로
mypath = './윤석열 댓글/'

# 모든 댓글을 저장할 리스트
all_comments = []

# 파일 목록 가져오기
onlyfiles = [f for f in os.listdir(mypath) if os.path.isfile(os.path.join(mypath, f))]
onlyfiles.sort()  # 파일 이름 정렬

# 파일 읽기 및 댓글 수집
for file in onlyfiles:
    file_path = os.path.join(mypath, file)
    with open(file_path, 'r', encoding='utf8') as f:
        content = f.readlines()  # 파일 내용을 줄 단위로 읽음
    all_comments.extend([line.strip() for line in content])  # 모든 댓글을 추가

# 랜덤 샘플링 (최대 10,000개)
max_comments = 30000
total_comments = random.sample(all_comments, min(max_comments, len(all_comments)))

# 결과 출력
print(f"총 수집된 댓글 수: {len(total_comments)}")
if total_comments:
    print(f"첫 번째 댓글: {total_comments[0]}")
    print(f"마지막 댓글: {total_comments[-1]}")
```

```python
len(total_comments)
```

```python
filtered_comments = [Kr_preprocessing(doc) for doc in total_comments]
```

```python
from itertools import chain
filtered_comments_one = list(chain.from_iterable(filtered_comments))
```

```python
from collections import Counter
best = Counter(filtered_comments_one)
print(best.most_common(10))
```

```python
import matplotlib.pyplot as plt

# 상위 20개의 단어와 빈도만 사용
top_n = 20  # 표시할 단어 수
words, counts = zip(*best.most_common(top_n))  # 단어와 빈도를 분리

# 막대그래프 생성
plt.figure(figsize=(12, 6))
plt.barh(words, counts, color='#E61E2B', alpha=0.8)  # 수평 막대그래프
plt.xlabel('빈도수', fontsize=12)
plt.ylabel('단어', fontsize=12)
plt.title('상위 20 댓글 빈도 분석 (검색어: 윤석열)', fontsize=16)
plt.gca().invert_yaxis()  # y축 뒤집기 (가장 빈도가 높은 단어가 위로 오도록)

# 그래프 출력 및 저장
plt.tight_layout()
output_file = 'ysy_top_20_comments_analysis.png'  # 저장할 파일 이름
plt.savefig(output_file, dpi=300, bbox_inches='tight')  # DPI와 여백 설정
plt.show()
```
