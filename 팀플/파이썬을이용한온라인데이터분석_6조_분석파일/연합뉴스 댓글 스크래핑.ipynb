# 1. 윤석열 

```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_experimental_option("detach", True)
chrome_options.add_argument("--disable-notifications")  # 알림 비활성화
#chrome_options.add_argument("--start-maximized")  # 브라우저 최대화
chrome_options.add_argument("--disable-popup-blocking")  # 팝업 차단 비활성화
chrome_options.add_argument("--headless")  # 셀레늄 창 보이지 않기

# CSV 파일 경로
csv_file = "윤석열_연합뉴스_기사목록200.csv"

# Selenium WebDriver 설정
s = Service(r"C:\Users\Administrator\2024-2\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=s, options=chrome_options)
driver.set_page_load_timeout(60)

# CSV 파일 로드
df = pd.read_csv(csv_file)

# 댓글 저장용 텍스트 파일 경로
output_dir = "output"
os.makedirs(output_dir, exist_ok=True)
comments_file_path = os.path.join(output_dir, "Yonhap_YSY_Comments.txt")

# 댓글 크롤링 함수
def scrape_comments_selenium(url):
    try:
        driver.get(url)

        # 댓글 섹션 로드 대기 (최대 10초)
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "reply-content"))
        )

        # 페이지 소스 가져오기
        soup = BeautifulSoup(driver.page_source, "lxml")

        # 댓글 섹션 탐색
        comments_section = soup.find_all("div", class_="reply-content-wrapper")
        if not comments_section:
            return []

        # data-content 속성에서 댓글 추출
        comments = [
            comment["data-content"].strip()
            for comment in comments_section if "data-content" in comment.attrs
        ]
        return comments

    except Exception as e:
        print(f"댓글 크롤링 실패: {url}, 오류: {e}")
        return []

# 크롤링 및 텍스트 파일로 저장
with open(comments_file_path, "a", encoding="utf-8") as comments_file:
    for idx, row in df.iterrows():
        url = row['링크']
        print(f"크롤링 중 ({idx + 1}/{len(df)}): {url}")

        # 댓글 크롤링
        comments = scrape_comments_selenium(url)

        if comments:
            print(f"댓글 저장 완료: 기사 {idx + 1}")
            for comment in comments:
                comments_file.write(f"1. {comment}\n")
            comments_file.write("\n")  # 기사별로 구분
        else:
            print(f"댓글이 비어 있습니다: 기사 {idx + 1}")

# WebDriver 종료
driver.quit()
```
