```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
chrome_options.add_argument("--disable-notifications")  # 알림 비활성화
chrome_options.add_argument("--start-maximized")  # 브라우저 최대화
chrome_options.add_argument("--disable-popup-blocking")  # 팝업 차단 비활성화

# ChromeDriver 경로 설정 및 Service 객체 생성
s = Service(r"C:\Users\Administrator\2024-2\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=s, options=chrome_options)

# 데이터 저장 리스트
titles, links = [], []

# 유튜브 검색 URL
url_Y = "https://www.youtube.com/@MBCNEWS11/search?query=%EC%9C%A4%EC%84%9D%EC%97%B4%20after%3A2022-02-14%20before%3A2022-03-09"
driver.get(url_Y)

# 페이지 로드 대기
time.sleep(5)

# 스크롤 다운 함수
def scroll_down(driver, pause_time=2, max_scroll=5):
    """페이지를 아래로 스크롤하여 더 많은 콘텐츠 로드"""
    last_height = driver.execute_script("return document.documentElement.scrollHeight")
    scroll_count = 0

    while scroll_count < max_scroll:
        # 스크롤 다운
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(pause_time)

        # 새로운 높이 가져오기
        new_height = driver.execute_script("return document.documentElement.scrollHeight")
        if new_height == last_height:  # 더 이상 콘텐츠가 로드되지 않음
            break

        last_height = new_height
        scroll_count += 1

# 스크롤 다운 수행
scroll_down(driver, pause_time=2, max_scroll=20)  # 스크롤 20번 또는 끝날 때까지

# BeautifulSoup 사용하여 HTML 파싱
page_source = driver.page_source
soup = BeautifulSoup(page_source, "lxml")

# 검색 결과의 각 동영상을 나타내는 태그 찾기
video_tags = soup.find_all("a", id="video-title")
for video in video_tags:
    try:
        # 제목 추출
        title = video.get("title", "제목 없음")
        titles.append(title)

        # 링크 추출
        link = f"https://www.youtube.com{video.get('href', '')}"
        links.append(link)

    except Exception as e:
        print(f"개별 동영상 처리 중 오류 발생: {e}")

# 데이터프레임 생성 및 저장
try:
    df = pd.DataFrame({"제목": titles, "링크": links})
    df.to_csv("MBC_윤석열_유튜브목록.csv", index=False, encoding="utf-8-sig")
    print("데이터가 CSV 파일로 저장되었습니다.")
except Exception as e:
    print("CSV 저장 중 오류 발생:", e)

# 브라우저 종료
driver.quit()
```
