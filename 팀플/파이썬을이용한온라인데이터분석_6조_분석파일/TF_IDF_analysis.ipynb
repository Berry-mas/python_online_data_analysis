```python
from kiwipiepy import Kiwi
import re
kiwi = Kiwi()
from kiwipiepy.utils import Stopwords
stopwords_dict = Stopwords()

# 사용자 정의 불용어를 키위 형태소 분석기의 불용어 사전에 추가하기
customized_stopwords = [
    '취재', '뉴스', '연합', '조선', '한겨레', '사전', '조사', '서울', '사진', '대표', '중앙',
    '유세', '더불어민주당', '국민의힘', '더불어 민주당', '국민의 힘', '허경영', '문재인', '박근혜',
    '원희룡', '윤석렬', '단일', '철수', '토론', '대선', '정권', '나라', '지지', '교체', '생각', '인간',
    '정치', '기사제보', '연합뉴스', '재배포', '사진구매', '조선일보', '한겨레', '토트넘', '후보',
    '안철수', '윤석열', '이재명', '기자', '대통령', '광고', '이날', '오전', '오후', '관련', '이번',
    '시작', '당시', '가능', '이후', '이상', '공개', '자신', '당선', '시간', '시절', '이야기', '전날',
    '저녁', '지난해', '지금', '민주당', '국힘', '국민', '투표', '선거', '대한민국', '포함', '정도',
    '말씀', '해당', '일정', '일부', '하루', '발언', '지난달', '미래', '가운데', '경우', '부분',
    '정부', '국민의당', '의원', '포인트', '공동', '국민의 당', '경북', '제주', '호남', '영남', '대구', '대전',
    '경기', '한국', '부산', '충청', '광주', '전북', '전남', '충남', '성남', '만큼', '경남', '경기도', '심상정', '정의당'
]
for word in customized_stopwords:
    stopwords_dict.add((word, 'NNG'))
    stopwords_dict.add((word, 'NNP'))

def Kr_preprocessing(text):
    text = re.sub(r'[^\d\s\w]','',text)
    kiwi_tokens = kiwi.tokenize(text, stopwords=stopwords_dict)
    noun_words=[]
    for token in kiwi_tokens:
        if 'NN' in token.tag and len(token.form)>1:
            noun_words.append(token.form)
    # 길이가 1인 단어 제거 하기
    final_noun_words = []
    for word in noun_words:
        if len(word)>1:
            final_noun_words.append(word)
    return final_noun_words
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# 파일 경로
file_path_lee = './ljm_youtube_comments.txt'  # 이재명 관련 파일
file_path_yoon = './ysy_youtube_comments.txt'  # 윤석열 관련 파일

# 파일 읽기 함수
def read_file(file_path):
    with open(file_path, 'r', encoding='utf8') as f:
        content = f.readlines()  # 줄 단위로 읽음
    return content

# 파일 데이터 읽기
lee_comments = read_file(file_path_lee)
yoon_comments = read_file(file_path_yoon)
```

```python
lee_filtered_docs = [Kr_preprocessing(content1) for content1 in lee_comments]
yoon_filtered_docs = [Kr_preprocessing(content2) for content2 in yoon_comments]
```

```python
lee_documents_filtered = []    ## 각각의 문서가 하나의 원소로 표현되어야 하기 때문에 join 함수를 통해 하나의 문자열로 묶어줘야 함
for doc1 in lee_filtered_docs:
    lee_document_filtered =' '.join(doc1)
    lee_documents_filtered.append(lee_document_filtered) # list of docs, doc 은 단어들로 구성이 된 string data
# 불필요한 단어들을 제거하고 난후 DTM로 변환하기 위해 다시 list of strings의 형태로 변환
```

```python
yoon_documents_filtered = []    ## 각각의 문서가 하나의 원소로 표현되어야 하기 때문에 join 함수를 통해 하나의 문자열로 묶어줘야 함
for doc2 in yoon_filtered_docs:
    yoon_document_filtered =' '.join(doc2)
    yoon_documents_filtered.append(yoon_document_filtered) # list of docs, doc 은 단어들로 구성이 된 string data
# 불필요한 단어들을 제거하고 난후 DTM로 변환하기 위해 다시 list of strings의 형태로 변환
```

```python
# TF-IDF Vectorizer 설정
tfidf_vectorizer = TfidfVectorizer(min_df=1, max_df=0.8, ngram_range=(1,1))

# 데이터 합치기
all_documents = lee_documents_filtered + yoon_documents_filtered

# TF-IDF 계산
tfidf_matrix = tfidf_vectorizer.fit_transform(all_documents)

# 단어 목록 가져오기
feature_names = tfidf_vectorizer.get_feature_names_out()

# 각 후보자에 해당하는 행렬 슬라이싱
lee_tfidf_matrix = tfidf_matrix[:len(lee_documents_filtered), :]
yoon_tfidf_matrix = tfidf_matrix[len(yoon_documents_filtered):, :]

# 후보자별 주요 키워드 추출 함수
def get_top_keywords(tfidf_matrix, feature_names, top_n=10):
    tfidf_mean = np.mean(tfidf_matrix.toarray(), axis=0)
    top_indices = np.argsort(tfidf_mean)[::-1][:top_n]
    top_keywords = [(feature_names[i], tfidf_mean[i]) for i in top_indices]
    return top_keywords

# 상위 키워드 추출
top_keywords_lee = get_top_keywords(lee_tfidf_matrix, feature_names, top_n=30)
top_keywords_yoon = get_top_keywords(yoon_tfidf_matrix, feature_names, top_n=30)

# 결과 출력
print("이재명 주요 키워드:")
for word, score in top_keywords_lee:
    print(f"{word}: {score:.4f}")

print("\n윤석열 주요 키워드:")
for word, score in top_keywords_yoon:
    print(f"{word}: {score:.4f}")

# 후보자별로 겹치는 키워드 확인
lee_keywords_set = set(word for word, score in top_keywords_lee)
yoon_keywords_set = set(word for word, score in top_keywords_yoon)

common_keywords = lee_keywords_set & yoon_keywords_set
print("\n공통 키워드:")
print(common_keywords)

unique_to_lee = lee_keywords_set - yoon_keywords_set
unique_to_yoon = yoon_keywords_set - lee_keywords_set
print("\n이재명 고유 키워드:")
print(unique_to_lee)
print("\n윤석열 고유 키워드:")
print(unique_to_yoon)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 한글 폰트 경로 지정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # Windows 사용자
# font_path = '/Library/Fonts/AppleGothic.ttf'  # macOS 사용자
# font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'  # Linux 사용자

# 워드클라우드 생성 함수
def generate_wordcloud(keywords, title, colormap, output_file):
    word_freq = {word: score for word, score in keywords}  # 단어와 TF-IDF 점수를 매핑
    wordcloud = WordCloud(
        font_path=font_path,
        width=800,
        height=400,
        background_color='white',
        colormap=colormap  # 색상 맵 지정
    ).generate_from_frequencies(word_freq)

    # 워드클라우드 출력
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.tight_layout()
    plt.savefig(output_file, dpi=300, bbox_inches='tight')  # 그래프 저장
    plt.show()  # 화면에 표시
    plt.close()  # 메모리 해제

# 이재명 워드클라우드 (파란 계열)
generate_wordcloud(
    top_keywords_lee,
    "이재명 주요 키워드 (TF-IDF)",
    colormap='Blues',
    output_file='lee_wordcloud.png'  # 저장 파일명
)

# 윤석열 워드클라우드 (빨간 계열)
generate_wordcloud(
    top_keywords_yoon,
    "윤석열 주요 키워드 (TF-IDF)",
    colormap='Reds',
    output_file='yoon_wordcloud.png'  # 저장 파일명
)
```
