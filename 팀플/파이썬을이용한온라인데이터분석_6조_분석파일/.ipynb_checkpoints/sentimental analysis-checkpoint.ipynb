```python
def Kr_preprocessing(text):
    # 특수문자 제거
    text = re.sub(r'[^\d\s\w]', '', text)
    # 형태소 분석 및 불용어 제거
    kiwi_tokens = kiwi.tokenize(text, stopwords=stopwords_dict)
    
    # 감정 관련 품사만 추출 (형용사, 동사, 명사 등)
    emotive_words = []
    for token in kiwi_tokens:
        if token.tag in ['VA', 'VV', 'MAG', 'NNG', 'NNP'] and len(token.form) > 1:
            emotive_words.append(token.form)
    return emotive_words
```

```python
import os

def preprocess_file(input_file, output_file):
    """
    텍스트 파일을 전처리하여 새로운 파일로 저장하는 함수.
    
    Args:
        input_file (str): 원본 텍스트 파일 경로
        output_file (str): 전처리된 텍스트 파일 저장 경로
    """
    try:
        # 파일 읽기
        with open(input_file, 'r', encoding='utf-8') as infile:
            lines = infile.readlines()
        
        processed_lines = []
        
        # 각 줄에 대해 전처리 수행
        for line in lines:
            processed_words = Kr_preprocessing(line.strip())  # 사용자 정의 전처리 함수 호출
            processed_line = ' '.join(processed_words)       # 단어를 공백으로 결합
            processed_lines.append(processed_line)
        
        # 전처리된 결과를 새로운 파일로 저장
        with open(output_file, 'w', encoding='utf-8') as outfile:
            outfile.write('\n'.join(processed_lines))
        
        print(f"파일 전처리 완료! 결과가 {output_file}에 저장되었습니다.")
    
    except Exception as e:
        print(f"오류 발생: {e}")

# 파일 경로 설정
input_file = 'ysy_youtube_comments.txt'  # 원본 파일 경로
output_file = 'ysy_youtube_comments_sentimental.txt'  # 저장할 파일 경로

# 함수 실행
preprocess_file(input_file, output_file)
```

# 2. 감성분석 및 학습 데이터 준비

```python
# 감성사전 기반, 긍/부정만

from kiwipiepy import Kiwi
import re

# 형태소 분석기 및 불용어 사전 로드
kiwi = Kiwi()

# 불용어 사전 설정 (사용자 정의 포함)
from kiwipiepy.utils import Stopwords
stopwords_dict = Stopwords()
customized_stopwords = [
    '취재', '뉴스', '연합', '조선', '한겨레', '사전', '조사', '서울', '사진', '대표', '중앙',
    '유세', '더불어민주당', '국민의힘', '더불어 민주당', '국민의 힘', '허경영', '문재인', '박근혜',
    '원희룡', '윤석렬', '단일', '철수', '토론', '대선', '정권', '나라', '지지', '교체', '생각', '인간',
    '정치', '기사제보', '연합뉴스', '재배포', '사진구매', '조선일보', '한겨레', '토트넘', '후보',
    '안철수', '윤석열', '이재명', '기자', '대통령', '광고', '이날', '오전', '오후', '관련', '이번',
    '시작', '당시', '가능', '이후', '이상', '공개', '자신', '당선', '시간', '시절', '이야기', '전날',
    '저녁', '지난해', '지금', '민주당', '국힘', '국민', '투표', '선거', '대한민국', '포함', '정도',
    '말씀', '해당', '일정', '일부', '하루', '발언', '지난달', '미래', '가운데', '경우', '부분',
    '정부', '국민의당', '의원', '포인트', '공동', '국민의 당', '경북', '제주', '호남', '영남', '대구', '대전',
    '경기', '한국', '부산', '충청', '광주', '전북', '전남', '충남', '성남', '만큼', '경남', '경기도', '심상정', '정의당'
]
for word in customized_stopwords:
    stopwords_dict.add((word, 'NNG'))
    stopwords_dict.add((word, 'NNP'))

def Kr_preprocessing(text):
    """
    댓글 전처리 함수: 형태소 분석 및 불용어 제거 포함.
    """
    text = re.sub(r'[^\d\s\w]', '', text)  # 특수 문자 제거
    kiwi_tokens = kiwi.tokenize(text, stopwords=stopwords_dict)  # 형태소 분석 및 불용어 제거
    noun_words = [token.form for token in kiwi_tokens if 'NN' in token.tag and len(token.form) > 1]  # 명사 추출
    return noun_words

def load_sentiment_dict(pos_file, neg_file):
    """
    감성 단어 사전 로드.
    """
    with open(pos_file, 'r', encoding='utf-8') as f:
        positive_words = set(line.strip() for line in f)
    with open(neg_file, 'r', encoding='utf-8') as f:
        negative_words = set(line.strip() for line in f)
    return positive_words, negative_words

def analyze_sentiments(input_file, positive_words, negative_words, output_file):
    """
    댓글 데이터를 감성 분석하여 긍정/부정만 결과 저장.
    """
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
        lines = infile.readlines()

        for line in lines:
            text = line.strip()  # 댓글 텍스트
            if not text:  # 빈 줄은 건너뜀
                continue
            
            tokens = Kr_preprocessing(text)  # 전처리된 단어 리스트

            # 감성 점수 계산
            pos_count = sum(1 for token in tokens if token in positive_words)
            neg_count = sum(1 for token in tokens if token in negative_words)
            sentiment_score = pos_count - neg_count

            # 감성 결과 (긍정: 1, 부정: 0)
            if sentiment_score > 0:
                sentiment = 1
            elif sentiment_score < 0:
                sentiment = 0
            else:
                sentiment = 0  # 중립도 부정으로 처리

            # 결과 저장 (댓글과 감성 결과만 기록)
            outfile.write(f"{text}\t{sentiment}\n")

        print(f"{input_file}에 대한 감성 분석 완료! 결과가 {output_file}에 저장되었습니다.")

# 감성 단어 사전 파일 경로
positive_words_file = 'positive_words.txt'
negative_words_file = 'negative_words.txt'

# 댓글 파일 경로
ysy_file = 'ysy_youtube_comments_sentimental.txt'
ijm_file = 'ljm_youtube_comments_sentimental.txt'

# 결과 파일 경로
ysy_output = 'ysy_sentiment_results_binary.txt'
ijm_output = 'ljm_sentiment_results_binary.txt'

# 감성 단어 사전 로드
positive_words, negative_words = load_sentiment_dict(positive_words_file, negative_words_file)

# 감성 분석 실행
analyze_sentiments(ysy_file, positive_words, negative_words, ysy_output)
analyze_sentiments(ijm_file, positive_words, negative_words, ijm_output)
```

```python
print(classification_report(test_labels, pred_labels_tf_l2))
```

```python
# Get coefficients of the model
coefficients = lr_tf_l2.coef_.tolist()

sorted_coefficients = sorted(enumerate(coefficients[0]), key=lambda x:x[1], reverse=True)
# 학습에 사용된 각 단어마다의 coefficient (즉 weight) 값이 존재
# coefficient값이 큰 순으로 정렬 'reverse=True'


K=30
# print top K positive words
print("긍정 단어 상위 {} 개".format(str(K)))
for word_id, coef in sorted_coefficients[:K]:
    print('{0:} ({1:.3f})'.format(words_dict[word_id], coef))
# print top K negative words
print("\n부정 단어 상위 {} 개".format(str(K)))
for word_id, coef in sorted_coefficients[-K:]:
    print('{0:} ({1:.3f})'.format(words_dict[word_id], coef))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 읽기
file_path = 'ysy_sentiment_results_binary.txt'
data = pd.read_csv(file_path, sep="\t", header=None, names=["Comment", "Sentiment"])

# 긍정/부정 비율 계산
sentiment_counts = data['Sentiment'].value_counts()
labels = ["Negative", "Positive"]
sizes = sentiment_counts.values
colors = ['#616161', '#64FFDA']  # 부정은 검은색, 긍정은 민트색

# 원 그래프 시각화
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '',
    colors=colors,
    startangle=90,  # 시작 각도를 90도로 설정
    shadow=False    # 그림자 제거로 더 깔끔하게
)
plt.title("ysy_sentiment Analysis Result", fontsize=14)  # 제목 크기 설정

# 그래프 저장
output_file = 'ysy_sentiment_rate.png'
plt.savefig(output_file, dpi=300, bbox_inches='tight')  # DPI와 여백 설정
plt.show()
```
