```python
import En_na
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
import nltk
from nltk.corpus import stopwords
import re
# 문장 단위 분석 (한 문서)
```

```python
# 텍스트 데이터를 읽어 옵니다.
with open('nytimes.txt', 'r', encoding='utf8') as f:
    content = f.read()
```

```python
selected_words = ['health', 'people', 'researcher', 'study', 'tie', 'interaction', 'friend', 'others', 'exercise', 'connection']
```

```python
g = En_na.do_na(content, selected_words)
nx.write_graphml(g, 'En_test.graphml')
nx.draw_networkx(g)
plt.show()
```

## 중심도 계산하기

```python
# 어떤 노드가 중요한, 중심적인 역할을 하는지 파악
```

```python
nx.degree_centrality(g)
```

```python
# 디그리 중심도
sorted(nx.degree_centrality(g).items(), key=lambda x:x[1],reverse=True)
# health와 people이 중요함
```

```python
# 매개 중심도
sorted(nx.betweenness_centrality(g).items(), key=lambda x:x[1],reverse=True)
```

```python
# 근접 중심도
sorted(nx.closeness_centrality(g).items(), key=lambda x:x[1],reverse=True)
```
