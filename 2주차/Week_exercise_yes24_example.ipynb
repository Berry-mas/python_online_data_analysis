```python
import requests
```

```python
url = 'https://www.yes24.com/Product/Goods/104663596'
```

```python
r = requests.get(url)  #r에 request 메시지를 저장
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup 
```

```python
soup = BeautifulSoup (r.text, 'lxml')
```

```python
soup.title.text
```

```python
soup.find('span', attrs={'class':'gd_data'}) #(왜 안되는지 확인해보기)
```

```python
soup.find('span', attrs={'class':'gd_data'}).text #(왜 안되는지 확인해보기)
```

```python
soup.find_all('class').text
```
