```python
from kiwipiepy import Kiwi
import re
```

```python
kiwi = Kiwi()
```

```python
with open('naver_news.txt', 'r', encoding='utf8') as f:
    content = f.read()
```

```python
print(content)
```

```python
# 기호 제거
filtered_content = re.sub(r'[^\s\w\d]',' ', content)
```

```python
print(filtered_content)
```

```python
# 형태소 분석
kiwi_tokens = kiwi.tokenize(filtered_content)
print(kiwi_tokens)
# Tag 정보
# https://github.com/bab2min/kiwipiepy#%ED%92%88%EC%82%AC-%ED%83%9C%EA%B7%B8  
```

```python
# 명사 단어/형태소 선택하기
Noun_words = []
for token in kiwi_tokens:
    if 'NN' in token.tag:
        Noun_words.append(token.form)
```

```python
print(Noun_words)
```

명사가 아닌 다른 품사의 단어들을 선택하고자 하는 경우는 아래와 같이 수행

```python
Featured_POS = ['NNG', 'NNP', 'NNB', 'VV', 'VA']
selected_words = []
for token in kiwi_tokens:
    if token.tag in Featured_POS:
        selected_words.append(token.form)
print(selected_words)
```

# 불용어 제거하기

### Kiwi 에서 제공되는 불용어 사전 사용

```python
from kiwipiepy.utils import Stopwords
stopwords_dict = Stopwords()
```

```python
stopwords_dict.stopwords
```

### 불용어 사전에 불용어 추가하기 <br>
두번째 종류의 불용어는 직접 사용자 정의 불용어 사전을 만들지 않고, <br>
kiwi 형태소 분석기에서 제공되는 불용어 사전에 불용어를 추가하여 제거한다. 

```python
# 한번에 하나의 단어 추가하기
stopwords_dict.add(('오늘', 'NNG'))
```

```python
# 한번에 여러 개의 단어 추가하기
customized_stopwords = ['연합뉴스', '네이버', '기자', '문의']
for word in customized_stopwords:
    stopwords_dict.add((word, 'NNG'))
```

### 불용어 제거하기

```python
# Kiwi를 이용해서 불용어 제거하기
kiwi.tokenize(filtered_content, stopwords=stopwords_dict)
```
